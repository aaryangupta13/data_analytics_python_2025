```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import ast
from datetime import datetime, date
from datasets import load_dataset

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
df_DA
```

```python
df_DA['job_posted_month_no'] = df['job_posted_date'].dt.month

df_DA
```

```python
df_DA_explode = df_DA.explode('job_skills')

df_DA_explode
```

```python
df_DA_pivot = df_DA_explode.pivot_table(index= 'job_posted_month_no', columns= 'job_skills', aggfunc= 'size', fill_value= 0)

df_DA_pivot
```

```python
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()

df_DA_pivot
```

```python
df_DA_pivot = df_DA_pivot.sort_values( "Total", axis= 1, ascending= False)

df_DA_pivot= df_DA_pivot.drop('Total')

df_DA_pivot
```

```python
df_DA_pivot.iloc[:,0:5].plot(kind= 'line')
plt.title(f"Top 5 skills for Data Analysts per month")
plt.xlabel('Month')
plt.ylabel('Count')

plt.show()
```

```python
df_DA_pivot = df_DA_pivot.reset_index()

df_DA_pivot['job_posted_month'] = pd.to_datetime(df_DA_pivot['job_posted_month_no'], format= '%m').dt.strftime('%b')

df_DA_pivot = df_DA_pivot.set_index('job_posted_month')

df_DA_pivot = df_DA_pivot.drop(['job_posted_month_no'], axis= 1)


df_DA_pivot.iloc[:,0:5].plot(kind= 'line')
plt.title(f"Top 5 skills for Data Analysts per month")
plt.xlabel('Month')
plt.ylabel('Count')

plt.show()
```
