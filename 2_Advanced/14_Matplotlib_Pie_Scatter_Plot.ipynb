```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import ast
from datetime import datetime, date
from datasets import load_dataset

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind= 'pie')
plt.title('Work from Home Status')
plt.ylabel("")

plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind= 'pie')
plt.title('Work from Home Status')
plt.ylabel("")

plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind= 'barh')
plt.title('Job Titles Status')
plt.ylabel("")
plt.gca().invert_yaxis()

plt.show()
```

```python
df['job_work_from_home'].value_counts().plot(kind= 'pie', startangle= 90, autopct= "%1.1f%%") 
# '%' says that we will be formatting the text from here onwards 
# '1' says the minimum digits before decimal places, to make it witha 1 decimal place we can say '1.1'
# 'f' specifies that this is a floating point number


plt.title('Work from Home Status')
plt.ylabel("")

plt.show()
```

```python
fig, ax = plt.subplots(1, 3)

dict_column = {
    'job_work_from_home': 'Work from Home',
    'job_no_degree_mention': 'Job Degree Req.',
    'job_health_insurance': 'Health Insurance Offered'
}

# ENUMERATE - For a list, it provides us a index back

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle= 90, autopct= "%1.1f%%", labels= ['False', 'True']) # NEED TO PUT LABELS IN A CORRECT ORDER *****
    ax[i].set_title(title)

fig.tight_layout()
plt.show()
```
