# How are in-demand skills trending for Data Analysts?

### Methodology

1. Aggregate skill counts monthly
2. Re-analyze based on percentage of total jobs
3. Plot the monthly skill demand

```python
# Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import ast
from datetime import datetime, date
from datasets import load_dataset

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title_short'] == "Data Analyst") & (df['job_country'] == "United States")].copy()
```

```python
df_DA_US["job_posted_month_no"] = df_DA_US["job_posted_date"].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode("job_skills")
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index= "job_posted_month_no", 
                             columns= "job_skills",
                             aggfunc= "size",
                             fill_value= 0)

df_DA_US_pivot.loc["Total"] = df_DA_US_pivot.sum()

df_DA_US_pivot = df_DA_US_pivot.sort_values("Total", axis= 1, ascending= False)

df_DA_US_pivot = df_DA_US_pivot.drop("Total")

df_DA_US_pivot
```

```python
df_DA_US_pivot.plot(kind= "line")
```

```python
df_DA_US_total = df_DA_US.groupby('job_posted_month_no').size()

df_DA_US_total
```

```python
df_DA_US_pivot
```

```python
df_DA_US_percent = df_DA_US_pivot.div(df_DA_US_total/100, axis= 0)

df_DA_US_percent
```

```python
df_DA_US_percent = df_DA_US_percent.reset_index()

df_DA_US_percent["job_posted_month"] = pd.to_datetime(df_DA_US_percent["job_posted_month_no"], format= '%m').dt.strftime("%b")

df_DA_US_percent = df_DA_US_percent.set_index("job_posted_month")

df_DA_US_percent = df_DA_US_percent.drop(columns = "job_posted_month_no")

df_DA_US_percent
```

```python
df_plot = df_DA_US_percent.iloc[:, :5]

sns.set_theme(style= "ticks")
sns.lineplot(data= df_plot, dashes= False, palette= "tab10")
sns.despine()
plt.title("Trending top skills for Data Analysts in the US")
plt.xlabel("2023")
plt.ylabel("Likelihood in Job Posting")

plt.legend().remove()

for i in range(5):
    if df_plot.columns[i] == "python":
        plt.text(len(df_plot.index)-.9, df_plot.iloc[-1, i]-0.7, df_plot.columns[i])
    elif df_plot.columns[i] == "tableau":
        plt.text(len(df_plot.index)-.9, df_plot.iloc[-1, i]+0.7, df_plot.columns[i])
    else:
        plt.text(len(df_plot.index)-.9, df_plot.iloc[-1, i], df_plot.columns[i])

from matplotlib.ticker import PercentFormatter
ax= plt.gca()
# ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f"{y:.0f}%"))
ax.yaxis.set_major_formatter(PercentFormatter(decimals= 0))

# plt.tight_layout()
plt.show()
```
